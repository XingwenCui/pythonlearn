```python
class Dog(object):
    
    def __init__(self,name):
        self.name = name
        
    def game(self):
        print("%s play jump"%self.name)
        
class XiaoTianDog(Dog):
    
    def game(self):
        print("%s fly in the sky"%self.name)
        
class Person(object):
    
    def __init__(self,name):
        self.name = name
        
    def game_with_dog(self,dog):
        print("%s and %s play happily"%(self.name,dog.name))
        
        dog.game()
        
wangcai = Dog("")

xiaotian = XiaoTianDog("")

xiaoming = Person("")

xiaoming.game_with_dog(wangcai)
xiaoming.game_with_dog(xiaotian)
```
