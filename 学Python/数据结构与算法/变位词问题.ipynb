# “变位词”判断问题
### ❖问题描述
所谓“变位词”是指两个词之间存在组成字母的
重新排列关系
如：heart和earth，python和typhon
为了简单起见，假设参与判断的两个词仅由小写
字母构成，而且长度相等
### ❖解题目标：写一个bool函数，以两个词作
为参数，返回这两个词是否变位词
### ❖可以很好展示同一问题的不同数量级算法





```python
import time
def findsu(n):
    start = time.perf_counter()
    for i in range(2,n):
        for j in range(2,i):
            if i % j == 0:
                break
            else:
                print(i)
    
    end = time.perf_counter()
    timing = end - start
    print("运行时间：{:.7f}s".format(timing))
    
findsu(100)
```

```python
import time
def anagramSolution1(s1,s2):
    start = time.perf_counter()
    alist = list(s2)
    pos1 = 0
    stillOK = True
    while pos1 < len(s1) and stillOK:
        pos2 = 0
        found = False
        while pos2 < len(alist) and not found:
            if s1[pos1] == alist[pos2]:
                found = True
            else:
                pos2 = pos2 + 1
        if found:
            alist[pos2] = None
        else:
            stillOK = False
        pos1 = pos1 + 1
    end = time.perf_counter()
    timing = end - start
    print("运行时间：{:.7f}s".format(timing))
    return stillOK
```

```python
anagramSolution1('abcdefghijklmnopqrstuvwxyz','zyxwvutsrqponmlkjihgfedcba')

```

## 解法2：排序比较

```python
def anagramSolution2(s1,s2):
    start = time.perf_counter()
    alist1 = list(s1)
    alist2 = list(s2)
    
    alist1.sort()
    alist2.sort()
    pos = 0
    matches = True
    while pos < len(s1) and matches:
        if alist1[pos] == alist2[pos]:
            pos = pos + 1
        else:
            matches = False
    end = time.perf_counter()
    timing = end - start
    print("运行时间：{:.7f}s".format(timing))    
    return matches
        
```

```python
anagramSolution2('abcdefghijklmnopqrstuvwxyz','zyxwvutsrqponmlkjihgfedcba')
```
