```python
class Bank:
    # creates a constructor for Bank
    def __init__(self, acc_name, acc_no, open_balance, address):
        self.acc_name = acc_name #represent the name of account holder
        self.acc_no = acc_no #indicate account number of the account holder
        self.balance = open_balance #indicate the balance of the account holder
        self.address = address # stores address of the account holder

    # returns the balance of account holder
    def check_bal(self):
        return self.balance
    
    # function performs withdraw from the account holder.
    def withdraw(self, amount):
        if amount > self.balance:
            return False
        else:
            self.balance -= amount

    # function performs change of account name for the account holder.
    def change_acc_name(self, new_name):
        self.acc_name = new_name

    # function performs transaction between 2 accounts
    def transaction(self, other, amount):
        if amount >self.balance:
            return False
        else:
            other.balance += amount
            self.balance -= amount

        # function writes account holders details into file
    def passBook(self, filename):
        with open(filename,'w',encoding='utf-8') as f:
            f.write("Holder's name is {},Holder's account number is {},Holder's balance is {},Holder's address is {}".format(
            self.acc_name,self.acc_no,self.balance,self.address))
            
    def __str__(self):
        return "hello"
```

```python
cxw = Bank("cxw","123456",1000,"")
zsh = Bank("zsh",'98765',500,"")
cxw.check_bal()
cxw.transaction(zsh,100)
cxw.check_bal(),zsh.check_bal()
cxw.passBook("cxw's passbook.txt")
print(cxw)
```

```python
def find_student_grade(inputFile1,inputFile2,outputFile):
    with open(inputFile1,'r') as f:
        file1_list = f.readlines()
    with open(inputFile2,'r') as f:
        file2_list = f.readlines()
    unit1_name = inputFile1.replace(".txt",'')
    unit2_name = inputFile2.replace(".txt",'')
    file1_dic = dict([i.strip().split(' ') for i in file1_list])
    file2_dic = dict([i.strip().split(' ') for i in file2_list])
    file_list = sorted(list(set(file1_dic) | set(file2_dic)))
    
    with open(outputFile,'a') as f:
        a_list = []
        for i in file_list:
            if i in file1_dic and i in file2_dic:
                avar_score = (float(file1_dic[i])+float(file2_dic[i]))/2
                information = i+','+unit1_name+','+unit2_name+','+str(avar_score)
                a_list.append(information)
            elif i in file1_dic:
                information = i+','+unit1_name+','+file1_dic[i]
                a_list.append(information)
            else:
                information = i+','+unit2_name+','+file2_dic[i]
                a_list.append(information)
        f.write("\n".join(a_list))
                
                
    
    
```

```python
find_student_grade("FIT9133.txt","FIT9136.txt","grade.txt")
```
