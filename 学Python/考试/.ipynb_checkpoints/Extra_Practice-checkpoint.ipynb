## 2. Simple Plagiarism Checker(Optional)

We will build a simple plagiarism checker in this task. The algorithm follows the description below:

1. Find the longest common substring between two strings(case sensitive).
2. The plagiarism score is calculated based on $\frac{length_{LongestCommonSubstring}}{length_{String1} + length_{String2} - length_{LongestCommonSubstring}}$, where <small>$length_{LongestCommonSubstring}$</small> is the length of the longest common substring between two strings in characters, <small>$length_{String1}$</small> is the length in characters of the first string and <small>$length_{String2}$</small> is the length in characters of the second string.
3. If both strings are empty strings, then the plagiarism score will be `1.0`.

Please write a piece of code that asks user to enter two strings separately and prints out the plagiarism score *rounded to* 1 decimal point.

<pre>
<b>Example 1:</b>
Input:
<i>This is Sentence 1.
This is Sentence 2.</i>
Output:
<i>The plagiarism score is 0.8</i>
</pre>
