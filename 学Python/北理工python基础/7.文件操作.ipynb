## 7.1文件的使用
### 7.1.1文件的类型
`文件是数据的抽象和集合`
- 文件是存储在辅助存储器上的数据序列
- 文件是数据存储的一种形式
- 文件展现形态：文本文件和二进制文件，本质上所有文件都是二进制形式存储，有统一编码的叫文本文件，没有的叫二进制文件，都能通过二进制打开。

`文本文件`是数据的抽象和集合
- 由**单一特定编码**组成的文件，如UTF-8编码
- 由于存在编码，也被看成是存储着的长字符串
- 适用于例如：txt文件、.py文件  

`二进制文件`直接由比特0和1组成，最原始的存储格式，**没有统一编码**，一般存在二进制0和1的组织结构，即文件格式，比如png文件、.avi文件

#### 文件打开模式
| 文件的打开模式 | 描述 | 
| :-----| :---- | 
| 'r' | 只读模式，默认值，如果文件不存在，返回FileNotFoundError |
| 'w' | 覆盖写模式，文件不存在则创建，存在则完全覆盖 |
| 'x' | 创建写模式，文件不存在则创建，存在则返回FileExistsError | 
| 'a' | 追加写模式，文件不存在则创建，存在则在文件最后追加内容 |
| 'b' | 二进制文件模式 |
| 't' | 文本文件模式，默认值 |
| '+' | 与r/x/w/a一同使用，在原功能基础上增加同时读写能力 |

````python
f = open("f.txt","w")   #文本形式、覆盖写模式，若文件不存在将创建文件
f = open("f.txt")       #文本形式、只读模式、默认值
f = open("f.txt","rt")  #文本形式、只读模式、默认值
f = open("f.txt","a+")  #文本形式、追加写模式+读文件内容
f.close()   #最后记得关文件，但是程序退出也会自动关
````

#### 文件的全文本操作
遍历全文本：方法一，一次读入，统一处理

````python
fname = input("please enter the file's name:")
fo = open(fname,'r',encoding = 'utf-8')
txt = fo.read()  #直接对全文处理,保存为一个字符串txt
fo.close()
# print(txt)
````

遍历全文本：方法二，按数量读入，逐步处理

````python
fname = input("please enter the file's name:")
fo = open(fname,'r',encoding = 'utf-8')
txt = fo.read(2)
while txt != '':   #对txt进行处理
    txt = fo.read(2)  #这里每次读入2个字符，直到字符为空，停止读入

fo.close()
````

#### 文件的逐行操作
逐行遍历文件：方法一，一次读入，分行处理

````python
fname = input("please enter the file's name:")
fo = open(fname,'r',encoding = 'utf-8')
for line in fo.readlines():   #读入文件所有行，每行为元素生成列表
    print(line)
fo.close()
````

````python
逐行遍历文件：方法二。分行读入逐行处理
fname = input("文件名称")
fo = open(fname,"r",encoding = 'utf-8')
for line in fo.readlines():    #readlines会将文本所有行读入一个列表里
    print(line)     #遍历打印列表里的每一个信息
fo.close()
````

## 7.3 一维数据处理
从一个数据到一组数据，一个数据（3.14）表达一个含义，一组数据(2,1,2,4)表达一个或多个含义  
**一维数据：是由对等关系的有序或无序数据构成，采用线性方式组织 -对应了列表、数组和集合等概念**  
二维数据：由多个一维数据构成，是一维数据的组合形式 -表格就是典型的二维数据，其中表头是二维数据的一部分  
多维数据：由一维或二维数据在新维度上扩展形成的，比如不同年度的大学排行榜。是由不同时间上的二维数据构成  
高维数据：仅利用最基本的二元关系展示数据间的复杂结构，如键值对
```python
{
    "name":"tian",
    "lastName":"song",
    "address":{
            "city":"beijing",
            "zipcode":"100081"
           }
    "professional":["Computer Networking","Security"]
}
```
#### 数据的操作周期

#### 一维数据的表示
如果数据间有序：使用列表类型
如果数据间无序：使用集合类型
- for循环可以遍历数据，进而对每个数据进行处理

#### 一维数据的存储
存储方式一：空格分隔（中国 美国 日本 意大利）
- 使用一个或多个空格分隔存储，不换行
- 缺点：数据中不能存在空格，否则会辨别错误

存储方式二：逗号分隔（中国,美国,德国）
- 使用英文半角逗号分隔数据进行存储，不换行
- 缺点：数据中不能有英文逗号

存储方式三：其他方式，如（中国@美国@德国）
- 使用其他符号或符号组合分隔，建议采用特殊符号
- 缺点：需要根据数据特点定义（即不会在数据中出现的），通用性差

#### 一维数据的处理


````python
#一维数据写入
#采用空格分隔方式将数据写入文件
ls = ['中国','美国','日本','德国']
f = open('fname.txt','w')
f.write(' '.join(ls))  #join是将join前面的字符串，分隔放置到后边的参数中
f.close()
````

````python
#特殊分隔方式
ls = ['中国','美国','日本','德国']
f = open('fname.txt','w')
f.write('$'.join(ls))  #join是将join前面的字符串，分隔放置到后边的参数中
f.close()
````

````python
#一维数据读入
#从空格分隔文件中读入数据
txt = open("fname.txt").read()
ls = txt.split() #以空格分隔
f.close()
ls
````

````python
# 从特殊符号分隔的文件中读入数据
txt = open("fname.txt").read()
ls = txt.split("$")
f.close()
ls
````

### CSV格式与二维数据存储
CSV: Comma-Separated Values（用逗号分隔值得方式）
- 国际通用的一二维数据存储格式，一般.csv扩展名
- 每行一个一维数据，采用逗号分隔，无空行
- Excel和一般编辑软件都可以读入或另存为csv文件
#### CSV数据存储格式

- 如果某个元素缺失，逗号仍要保留
- 二维数据的表头可以作为数据存储，也可以另行存储
- 逗号为英文半角逗号，逗号与数据之间无额外空格
- 一般索引习惯是:ls[row][column],先行后列
- 根据一般习惯，外层列表每个元素是一行，按行存
