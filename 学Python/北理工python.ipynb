摄氏度华氏度相互转换\
问题分析：
- 输入：带华氏度摄氏度标志的温度值
- 处理：根据温度标志选择适当的温度转换算法
- 输出：带摄氏度或华氏度标志的温度值

````python
#TempConvert.py
TempStr = input("请输入带有符号的温度值：")
if TempStr[-1] in ['F','f']:
    C = (eval(TempStr[0:-1]) - 32)/1.8
    print("转换后的温度是{:.2f}C".format(C))
elif TempStr[-1] in ['C','c']:
    F = 1.8*eval(TempStr[0:-1]) + 32
    print("转换后的温度是{:.2f}F".format(F))
else:
    print("输入格式错误")
````

## 2.3 turtle库概述
turtle（海龟）库是turtle绘图体系的python实现
`Python计算生态 = 标准库 + 第三方库`
- 标准库：随解释器直接安装到操作系统中的功能模块
- 第三方库：需要经过安装才能使用的功能模块
- 库Library、包package、模块module，统称模块
turtle的原理是假设turtle是一个真实存在的，有一只海龟在窗体画布上游走，走过的轨迹就是画图  
turtle包括：
1. 绘图窗体：  
turtle的一个画布空间，最小单位是像素，用turtle.setyp(width,height,startx,starty)设置窗体大小位置,后两个参数是可选参数以电脑屏幕的右上角为（0,0）原点
2. 空间坐标体系：

3. turtle角度坐标体系：  
turtle.seth(angle)改变海龟行进方向，但不行进，angle为绝对度数即绝对坐标系下的角度，配合fd会使海龟按当前方向直线行驶  
turtle.left(),turtle.right()是海龟视角下的左右的角度
4. RGB色彩体系
由三种颜色组成的万物色，turtle.colormode（mode）改变颜色
5. 画笔控制函数
turtle.penup()抬起画笔
turtle.pendown()放下画笔
turtle.pensize()画笔大小
turtle.pencolor(color)笔的颜色

## 遍历循环
遍历**某个结构形成**的循环运行方式
```
    for  <循环变量>  in  <遍历结构> :
        <语句块>
```
从遍历结构中逐一提取元素，放在循环变量中，完整遍历所有元素后结束，每次循环所获得的元素放入循环变量中并执行一次语句块。  
有以下应用方式：  
- 计数循环（N次）
``` 
for i in range(M,N,K):   从M开始，到N-1，以K为步长，i是变量。
    <语句块>
```

## 循环控制保留字
    break 和 continue
- break跳出并结束当前整个循环，执行循环后的语句
- continue结束当次循环，继续执行后续次数循环，即*并不打破循环只是打破当次循环*
- 都可与for，while 循环搭配使用

````python
for c in "PYTHON":
    if c == "T":
        continue   #在执行循环时，如果当次循环遇到了continue那当次循环就跳过，继续执行下边的
    print(c,end="")
print(" ")        #换行
for c in "PYTHON":
    if c == "T":
        break   #在执行循环时，如果当次循环遇到了break，就跳出当前循环
    print(c,end="")
````

````python
s ="PYTHON"
while s !="":
    for c in s:
        print(c,end="")
    s = s[:-1]
print(" ")
s ="PYTHON"
while s !="":
    for c in s:
        if c == "T":
            break   #仅跳出最内循环
        print(c,end="")
    s = s[:-1]
````

## 实例圆周率的计算
### 问题分析
- 数学思维：找到公式，利用公式求解

- 计算机思维：抽象一种过程，用计算机自动化求解(运用蒙特卡洛方法，对一个区域进行撒点，计算频率)

````python
#公式法
pi = 0
N = 100
for k in range(N):
    pi += 1/pow(16,k)*(4/(8*k+1)-2/(8*k+4)-\
          1/(8*k+5)-1/(8*k+6))
print("圆周率是:{}".format(pi))
````

````python
#蒙特卡洛法,模拟1000*1000次撒点
from random import random     #调用random库获得random函数,获得[0,1)内的随机数
from time import perf_counter   #使用time库的计时方式获得程序的运行时间
DARTS = 1000*1000                #撒点数量
hits = 0.0           #在圆的内部点的数量
start = perf_counter()     #启动计时
for i in range(1,DARTS+1):     #模拟撒点，循环1000*1000次
    x,y = random(),random()     #设成（x,y）坐标，范围是[1,1]
    dist = pow(x**2 + y**2, 0.5)  #计算distinction距离，即以（0,0）为圆心的半径
    if dist <= 1.0:               #如果半径不大于1，则在圆内
        hits = hits + 1
pi = 4 * (hits/DARTS)            #模拟的是1/4圆，×4就是整个圆
print("圆周率的值是:{}".format(pi))
print("运行时间是：{:.5f}s".format(perf_counter()-start))   #时间主要花在了循环上
````

# 5 函数和代码的复用
## 5.1函数的定义和使用
函数定义一段代码的表示：
```python
    def <function_name>(<parameter(0个或多个)>):      参数是一种占位符
        <函数体>
        return <返回值>
```
- 函数是一段具有特定功能的、可重用的语句组,一段完整的代码封装
- 函数是一种功能的抽象，一般函数表达特定功能
- 两个作用：降低编程难度和代码复用，定义了以后直接用不用管内部逻辑，并且可以重复调用
- 函数定义后，如果不经调用，不会被执行
- 函数定义时，参数是输入，函数体是处理、结果是输出（IPO）


````python
#计算n！
def fact(n):
    s = 1
    for i in range(1,n+1):
        s *= i
    return s

fact(3)       #函数的调用
````

### 函数的参数传递
`函数可以有参数也可以没有但必须要保留括号`  
函数定义时可以为某些参数指定默认值，构成可选参数
```python
    def <函数名>（<非可选参数>,<可选参数>）:   非可选参数又叫必须参数，一定要放在前边
        <函数体>
        return<返回值>

```
`函数定义时可以设计可变数量的参数，即不确定参数总数量`
```python
    def <name>(<参数>,*b):       *b表示不确定的参数
        <函数体>
        return <value>
```
参数传递指的是在函数调用时参数可以按照位置或名称的方式传递，即fact（10,5）或fact（m=10，n=5）

````python
def fact(n,*b):
    s = 1
    for i in range(1,n+1):
        s *= i
    for item in b:
        s *= item
    return s

print(fact(10,3))
print(fact(10,3,2))  #可以一个也可以两个参数
print(fact(n = 10))
````

### 函数的返回值
函数可以返回0个或多个结果  
- return关键字用来传递返回值  
- 函数可以有返回值，也可以没有，可以有return，也可以没有，return可以传递多个返回值

````python
def fact(n,m=1):
    s = 1
    for i in range(1,n+1):
        s *= i
    return s//m, n, m
fact(10,5)
print(fact(10,5))
a,b,c = fact(10,5)
print(a,b,c)
````

### 局部变量全局变量
局部是函数内部有使用的变量，函数外部整个程序使用的变量叫全局变量
- `局部变量和全局变量是不同的变量`  
局部变量是函数的内部占位符，与全局变量可能重名但不同  
函数运算结束后，局部变量被释放，就是后边没用了，只是在函数内部构造函数用的  
可以使用global关键字在函数内部使用全局变量
- `局部变量为组合数据类型且未创建，等同于全局变量`  
比如列表，如果不是在函数内部创建，而是本来就存在的，那么内部的也算全局变量，"指针"

````python
n,s = 10, 100
def fact(n):
    s = 1                 #这里的s只是当前函数局部变量
    for i in range(1,n+1):
        s *= i
    return s
print(fact(n),s)          #输出的是函数返回值和全局变量s
````

````python
n,s = 10, 100
def fact(n):
    global s                 #函数中使用global保留字声明此处s是全局变量s
    for i in range(1,n+1):
        s *= i
    return s             #此处指的是全局变量s
print(fact(n),s)          #输出的是函数返回值和全局变量s
````

````python
ls = ["F","f"]     #通过使用[]真实创建了一个全局变量列表ls
def func(a):       
    ls.append(a)   #此时ls是列表类型，未真实创建则等同于全局变量
    return
func("C")          #全局变量ls被修改
print(ls)
````

````python
ls = ["F","f"]     #通过使用[]真实创建了一个全局变量列表ls
def func(a):       
    ls = []        #真实创建了同名ls
    ls.append(a)   #此时ls是列表类型，真实创建则是局部变量，是对局部变量的修改
    return 
func("C")          #局部变量ls被修改
print(ls)
````

### 实例7 七段数码管绘制
七段数码管是七个数码管绘制的数字，通过控制数码管亮度获得0-9,a-f的图形，如红绿灯，可以用turtle库来绘制  
**基本思路**  
- 步骤一：绘制单个数字的数码管
- 步骤二：获得一串数字，绘制对应的数码管
- 步骤三：获得当前系统时间，绘制对应的数码管

````python
import turtle as t
import time
def drawGap():
    t.penup()
    t.fd(5)
def drawLine(draw):    #绘制单段数码管
    drawGap()
    t.pendown() if draw else t.penup()
    t.fd(40)
    drawGap()
    t.right(90)
def drawDigit(digit):  #根据数字绘制七段数码管
    drawLine(True) if digit in [2,3,4,5,6,8,9] else drawLine(False)
    drawLine(True) if digit in [0,1,3,4,5,6,7,8,9] else drawLine(False)
    drawLine(True) if digit in [0,2,3,5,6,8,9] else drawLine(False)
    drawLine(True) if digit in [0,2,6,8] else drawLine(False)
    t.left(90)
    drawLine(True) if digit in [0,4,5,6,8,9] else drawLine(False)
    drawLine(True) if digit in [0,2,3,5,6,7,8,9] else drawLine(False)
    drawLine(True) if digit in [0,1,2,3,4,7,8,9] else drawLine(False)
    t.left(180)        #drawLine函数每次绘制完会向右转90度
    t.penup()          #为绘制后续数字确定位置
    t.fd(20)           #为绘制后续数字确定位置
def drawDate(date):       #date为日期，格式为'%X-%m=%d+'
    t.pencolor("red")
    for i in date:
        if i == '-':
            t.write("年",font=("Arial",18,"normal"))
            t.pencolor("green")
            t.fd(40)
        elif i == '=':
            t.write("月",font=("Arial",18,"normal"))
            t.pencolor('blue')
            t.fd(40)
        elif i == '+':
            t.write("日",font=("Arial",18,"normal"))
        else:
            drawDigit(eval(i))
def main():
    t.setup(800,350,200,200)
    t.penup()
    t.fd(-300)
    t.pensize(5)
    drawDate(time.strftime("%Y-%m=%d+",time.gmtime()))
    t.hideturtle()
    t.done()
main()
````

## 实例9：基本统计值
- 需求：给出一组数，对它有个概要理解
- 该怎么做？总个数、求和、均值、方差...\
 总个数：len()\
 求和：for ... in\
 平均值：求和/总个数\
 方差：各数据与平均数差的平方和的平均数\
 中位数：排序，然后奇数找中间一个，偶数找中间两个平均

````python
#获得用户的每一个数据输入，直到用户点回车
def getNum():
    nums = []
    iNumStr = input("请输入数字(回车键退出)：")
    while iNumStr != "":    #只要输入不是""空，即回车没有带任何数据就一直循环
        nums.append(eval(iNumStr))
        iNumStr = input("请输入数字(回车键退出)：")
    return nums

def mean(numbers):
    s = 0.0
    for num in numbers:
        s = s + num
    return s/len(numbers)

def dev(numbers,mean):
    sdev = 0.0
    for num in numbers:
        sdev = sdev + (num - mean)**2
    return pow(sdev/(len(numbers)-1),0.5)
def median(numbers):
    sorted(numbers)
    size = len(numbers)
    if size % 2 == 0:
        med = (numbers[size//2-1] + numbers[size//2])/2
    else:
        med = numbers[size//2]
    return med
n = getNum()
m = mean(n)
print("平均值：{},方差{:.2},中位数：{}.".format(m,dev(n,m),median(n)))
````

## 实例10 文本词频统计

````python
#CalHamlet
#数据清洗，对文本进行处理，把每个词都分开，即去掉所有符号替换成空格，将所有字母进行小写处理
def getText():
    txt = open("hamlet.txt","r").read()
    txt = txt.lower()
    for ch in '!"#$%&^_|{}`:;@''?/_-+=|\()':
        txt = txt.replace(ch," ")
    return txt

hamletTxt = getText()
words = hamletTxt.split()
counts = {}
for word in words:
    counts[word] = counts.get(word,0) + 1
items = list(counts.items())
items.sort(key=lambda x:x[1],reverse=True)
for i in range(10):
    word,count = items[i]
    print("{0:<10}{1:>5}".format(word,count))
````

## 7.1文件的使用
#### 文件的类型
`文件是数据的抽象和集合`
- 文件是存储在辅助存储器上的数据序列
- 文件是数据存储的一种形式
- 文件展现形态：文本文件和二进制文件，本质上所有文件都是二进制形式存储，有统一编码的叫文本文件，没有的叫二进制文件，都能通过二进制打开。  
`文本文件`是数据的抽象和集合
- 由**单一特定编码**组成的文件，如UTF-8编码
- 由于存在编码，也被看成是存储着的长字符串
- 适用于例如：txt文件、.py文件  
`二进制文件`直接由比特0和1组成，最原始的存储格式，**没有统一编码**，一般存在二进制0和1的组织结构，即文件格式，比如png文件、.avi文件
#### 文件的打开关闭

**文件处理步骤：打开-操作-关闭**

文件打开：程序在**运行的当前目录下**找到并打开这个文件
```python
<变量名> = open(<文件名>,<打开模式>)

- 变量名是一个抽象的变量，用文件句柄来表示
- 文件名是文件路径+名称，如果**源文件同目录可以省略路径**，同时路径中的\用/ 来代替(python中\是转义字符)
"D:/pye/f.txt"绝对路径
"./pye/f.txt"相对路径，该project路径都在D盘里下面
"f.txt"相同目录，直接打开
```

## 实例11.自动轨迹绘制
根据脚本绘制图形而不是使用代码，写程序读取数据，数据脚本是自动化的重要一步\
基本思路：
- 步骤1 定义数据文件格式（接口）
- 步骤2 编写数据，根据文件接口解析参数绘制图形
- 步骤3 编制数据文件

````python
#自动轨迹绘制
import turtle as t
t.title('自动轨迹绘制')   #基础准备工作
t.setup(800,600,0,0)
t.pencolor("red")
t.pensize(5)
#数据读取
datals = []   #建立空列表,数据读取之后以列表保存
f = open("data.txt")    #这里是一个脚本文件，里面有数据来驱动，设计接口方便读取
for line in f:          #把数据读取出来赋给line，数据接口的值，拿到的是字符串
    line = line.replace("\n","")   #replace将换行符转换成空字符串，去掉换行信息
    datals.append(list(map(eval,line.split(","))))   #map将第一个参数的功能作用于每一个元素，即都eval一次，split","以逗号分割返回列表
f.close()
#自动绘制
for i in range(len(datals)):
    t.pencolor(datals[i][3],datals[i][4],datals[i][5])
    t.fd(datals[i][0])
    if datals[i][1]:
        t.right(datals[i][2])
    else:
        t.left(datals[i][2])
````

## 7.2一维数据的处理
从一个数据到一组数据，一个数据（3.14）表达一个含义，一组数据(2,1,2,4)表达一个或多个含义  
**一维数据：是由对等关系的有序或无序数据构成，采用线性方式组织 -对应了列表、数组和集合等概念**  
二维数据：由多个一维数据构成，是一维数据的组合形式 -表格就是典型的二维数据，其中表头是二维数据的一部分  
多维数据：由一维或二维数据在新维度上扩展形成的，比如不同年度的大学排行榜。是由不同时间上的二维数据构成  
高维数据：仅利用最基本的二元关系展示数据间的复杂结构，如键值对
```python
{
    "name":"tian",
    "lastName":"song",
    "address":{
            "city":"beijing",
            "zipcode":"100081"
           }
    "professional":["Computer Networking","Security"]
}
```

#### 一维数据的表示
如果数据间有序：使用列表类型
如果数据间无序：使用集合类型
- for循环可以遍历数据，进而对每个数据进行处理

#### 一维数据的存储
存储方式一：空格分隔（中国 美国 日本 意大利）
- 使用一个或多个空格分隔存储，不换行
- 缺点：数据中不能存在空格，否则会辨别错误
存储方式二：逗号分隔（中国,美国,德国）
- 使用英文半角逗号分隔数据进行存储，不换行
- 缺点：数据中不能有英文逗号
存储方式三：其他方式，如（中国@美国@德国）
- 使用其他符号或符号组合分隔，建议采用特殊符号
- 缺点：需要根据数据特点定义（即不会在数据中出现的），通用性差

````python
#一维数据写入
#采用空格分隔方式将数据写入文件
ls = ['中国','美国','日本','德国']
f = open('fname.txt','w')
f.write(''.join(ls))  #join是将join前面的字符串，分隔放置到后边的参数中
f.close()
#特殊分隔方式
# ls = ['中国','美国','日本','德国']
# f = open('fname.txt','w')
# f.write('$'.join(ls))  #join是将join前面的字符串，分隔放置到后边的参数中
# f.close()
````
